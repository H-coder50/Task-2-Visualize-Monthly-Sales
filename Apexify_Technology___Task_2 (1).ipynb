# **Visualize Monthly Sales:**
## **1. Create Monthly Sales Data:**

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a dummy sales.csv file for demonstration
# -----------------------------------------------------
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales": [12000, 13500, 15000, 14500, 16000, 15500,
              17000, 17500, 16500, 18000, 19000, 20000]
}

df = pd.DataFrame(data)
df.to_csv('sales.csv', index=False)
df
```

## **2: Plot the Line Chart (Sales Trend):**

```python

# Style
plt.figure(figsize=(12,6))
plt.style.use('seaborn-v0_8-darkgrid')

# Line Plot
plt.plot(df["Month"], df["Sales"],
         marker='o',
         markersize=10,
         linewidth=3,
         linestyle='-',
         color="#007acc")   # Blue theme

# Highlight max sales point
max_month = df.loc[df["Sales"].idxmax(), "Month"]
max_sales = df["Sales"].max()
plt.scatter(max_month, max_sales, color="red", s=120, label="Highest Sales")

# Titles
plt.title(" Monthly Sales Trend ", fontsize=20, fontweight='bold')
plt.xlabel("Month", fontsize=14)
plt.ylabel("Sales Amount", fontsize=14)

# Customizations
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True, linestyle="--", alpha=0.5)

# Add value labels
for i, val in enumerate(df["Sales"]):
    plt.text(i, val + 500, str(val), ha='center', fontsize=10)

plt.legend()
plt.tight_layout()
plt.show()
```
